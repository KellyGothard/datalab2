### The Fraction of safe burials. 

```python
unique_chiefdoms = list(set(all_data_followup['Chiefdom']))[3:15]

# The fraction of all the burials that were safe compared to all burials 
# TODO: Use the cleaned dataset in order to shrink the number of district. 
#   - this would remove the .isin section of this code. Perhaps then just a filtering.
all_data_followup['burial_fraction'] = 1 - \
(all_data_followup.loc[all_data_followup['Chiefdom'].isin(unique_chiefdoms)].filter(like = 'cb_').sum(axis = 1) / \
 all_data_followup.loc[all_data_followup['Chiefdom'].isin(unique_chiefdoms)].filter(like = 'b_').sum(axis = 1))

g = sns.FacetGrid(all_data_followup.\
loc[all_data_followup['Chiefdom'].\
    isin(unique_chiefdoms)], col = 'Chiefdom', col_wrap = 4)

g = g.map(plt.scatter, 'follow_up_date', 'burial_fraction')
plt.tight_layout()
plt.figure(figsize = (10, 5))

# Rotate the X axis 
for ax in g.axes.ravel():
    ax.set_xticklabels(ax.get_xticklabels(), rotation=60)
```

### The fraction of Health Referrals. 

- <font color = 'red'>This makes the assumption that the fraction of refferels is out of the number of seriously sick cases. 

```python
all_data_followup['referred_fraction'] = 1 - \
(all_data_followup.loc[all_data_followup['Chiefdom'].isin(unique_chiefdoms)].filter(like = 'r_').sum(axis = 1) / \
 all_data_followup.loc[all_data_followup['Chiefdom'].isin(unique_chiefdoms)].filter(like = 'ss_').sum(axis = 1))

# all_data_followup['referred_fraction_mean'] = \
# all_data_followup.loc[all_data_followup['Chiefdom'].isin(unique_chiefdoms)].groupby('Chiefdom').mean()


g = sns.FacetGrid(all_data_followup.\
loc[all_data_followup['Chiefdom'].\
    isin(unique_chiefdoms)], col = 'Chiefdom', col_wrap = 4)

g = g.map(plt.scatter, 'follow_up_date', 'referred_fraction')
g = g.map(sns.lineplot, 'follow_up_date', 'referred_fraction', estimator = np.mean)
plt.tight_layout()
plt.figure(figsize = (10, 5))

# Rotate the X axis 
for ax in g.axes.ravel():
    ax.set_xticklabels(ax.get_xticklabels(), rotation=60)


#TODO: Bin the values so the lines come out smoother
```
